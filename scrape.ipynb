```python
df2 = pd.DataFrame()
df2["name"] = df["name"]
df2["id"] = df["id"]
df2["max_float"] = df["max_float"]
df2["min_float"] = df["min_float"]
df2["crates"] = df["crates"].apply(lambda x: [crate['id'] for crate in x] if isinstance(x, list) else x)
df2["rarity"] = df["rarity"].apply(lambda x: x['name'])
df2["weapon"] = df["weapon"].apply(lambda x: x['name'])
df2["collections"] = df["collections"].apply(lambda x: [crate['name'] for crate in x] if isinstance(x, list) else x)
```

```python
import requests
from urllib.parse import quote

def get_steam_market_listings(market_hash_name, app_id=730, start=0, count=10, 
                               country='US', language='english', currency=1, query='', filter_param=''):
    """
    Get Steam Market listings for an item
    
    Args:
        market_hash_name: Item name (e.g., "Five-SeveN | Hyper Beast (Field-Tested)")
        app_id: Steam App ID (730 for CS:GO/CS2)
        start: Starting index for results
        count: Number of results to return
        country: Country code
        language: Language
        currency: Currency ID (1 = USD)
        query: Search query
        filter_param: Filter (e.g., sticker, charm, patch)
    """
    encoded_name = quote(market_hash_name)
    
    url = f"https://steamcommunity.com/market/listings/{app_id}/{encoded_name}/render/"
    
    params = {
        'query': query,
        'start': start,
        'count': count,
        'country': country,
        'language': language,
        'currency': currency,
        'filter': filter_param
    }
    
    headers = {
        'Host': 'steamcommunity.com',
        'Origin': 'https://steamcommunity.com/',
        'Referer': f'https://steamcommunity.com/market/listings/{app_id}/{encoded_name}' + 
                   (f'?filter={filter_param}' if filter_param else '')
    }
    
    response = requests.get(url, params=params, headers=headers)
    response.raise_for_status()
    
    return response.json()

# Get listings for Five-SeveN | Hyper Beast (Field-Tested)
item_name = "R8 Revolver | Banana Cannon (Well-Worn)"
data = get_steam_market_listings(item_name, count=5)
data
```

```python
# Let's search for items with "Five-SeveN" to see the exact names
search_url = "https://steamcommunity.com/market/search/render/"
params = {
    'query': 'Five-SeveN Hyper Beast',
    'start': 0,
    'count': 10,
    'search_descriptions': 0,
    'sort_column': 'popular',
    'sort_dir': 'desc',
    'appid': 730,
    'norender': 1
}

headers = {
    'Host': 'steamcommunity.com',
    'Referer': 'https://steamcommunity.com/market/search?appid=730'
}

search_response = requests.get(search_url, params=params, headers=headers)
search_data = search_response.json()

# Extract item names
for item in search_data.get('results', []):
    print(item.get('hash_name'))
```
